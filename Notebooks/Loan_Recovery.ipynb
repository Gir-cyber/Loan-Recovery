```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Default_In_2yrs'])
y = df['Default_In_2yrs']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)

```

```python
y_train.value_counts()
```

```python
X_train['Age_cat'] = pd.cut(
    X_train['Age'],
    bins=[0, 25, 40, 60, 120],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)
X_train.head()
```

```python
age_median_map = (X_train.groupby('Age_cat')['Age'].median())

age_median_map
```

```python
X_train['Dependents_cat'] = pd.cut(
    X_train['Dependents'],
    bins=[-1, 1, 3, 6, 20],
    labels=['None', 'Small', 'Medium', 'Large']
)
```

```python
dep_median_map = (X_train.groupby('Dependents_cat')['Dependents'].median())

dep_median_map
```

```python
X_train['Dependents'] = X_train['Dependents'].fillna(
    X_train['Dependents_cat'].map(dep_median_map)
)

X_train['Dependents'] = X_train['Dependents'].fillna(
    X_train['Dependents'].median()
)
X_train.isna().sum()
```

```python
X_test['Age_cat'] = pd.cut(
    X_test['Age'],
    bins=[0, 25, 40, 60, 120],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)

X_test['Dependents_cat'] = pd.cut(
    X_test['Dependents'],
    bins=[-1, 1, 3, 6, 20],
    labels=['None', 'Small', 'Medium', 'Large']
)

dep_median_map = (X_test.groupby('Dependents_cat')['Dependents'].median())

X_test['Dependents'] = X_test['Dependents'].fillna(
    X_test['Dependents_cat'].map(dep_median_map)
)

X_test['Dependents'] = X_test['Dependents'].fillna(
    X_test['Dependents'].median()
)
```

```python
numerical_features = [
    "Age",
    "Dependents",
    "Monthly_Income",
    "Real_Estate_Loans",
    "Debt_IncomeRatio",
    "Credit_Utilization",
    "Open_Credit_Lines",
    "Past_Due30_59",
    "Past_Due60_89",
    "Past_Due90"
]

categorical_features = [
    "Age_cat",
    "Dependents_cat"
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, numerical_features),
        ("cat", cat_pipeline, categorical_features)
    ]
)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
model = RandomForestClassifier(n_estimators=400,max_depth=6,class_weight="balanced",random_state=42,)

pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", model)
])
```

```python
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```
